```python
import zipfile
import re
import csv

# 1. hwpx 내부 XML 읽기
with zipfile.ZipFile('/Users/judetak/Desktop/산공카르텔경진대회/서울시 대기측정소 높이(2025).hwpx') as z:
    content = z.read('Contents/section0.xml').decode('utf-8')

# 2. 네임스페이스 제거 (파싱 편의)
content = re.sub(r' xmlns[^"]*"[^"]*"', '', content)
content = re.sub(r'<[a-zA-Z]+:', '<', content)
content = re.sub(r'</[a-zA-Z]+:', '</', content)

# 3. 표의 셀(tc) 단위로 텍스트 추출
cells = re.findall(r'<tc\b[^>]*>(.*?)</tc>', content, re.DOTALL)

texts = []
for cell in cells:
    t = re.findall(r'<t[^>]*>(.*?)</t>', cell, re.DOTALL)
    combined = ' '.join(t).strip()
    combined = re.sub(r'  +', ' ', combined)  # 공백 정리
    texts.append(combined)

# 4. 4열 기준으로 행 구성 (표 1: 도시대기 25개소, 표 2: 도로변대기 15개소)
table1 = [texts[i:i+4] for i in range(0, 104, 4)]   # 헤더 포함 26행
table2 = [texts[i:i+4] for i in range(104, 168, 4)]  # 헤더 포함 16행

# 5. CSV 저장 (utf-8-sig: 엑셀에서 한글 깨짐 방지)
with open('서울시_대기측정소_높이_2025.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerow(['[도시대기 측정망 25개소]'])
    writer.writerows(table1)
    writer.writerow([])
    writer.writerow(['[도로변대기 측정망 15개소]'])
    writer.writerows(table2)

print("CSV 저장 완료!")
```
