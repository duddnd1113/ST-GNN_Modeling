```python
import pandas as pd
import geopandas as gpd

def clean_building_data(gdf):
    # 원본 데이터 보호를 위해 복사본 생성
    df = gdf.copy()
    
    # ---------------------------------------------------------
    # 1. 건축물면적(㎡), 높이(m)가 0이고 데이터생성변경일자가 NaN인 행 삭제
    # ---------------------------------------------------------
    drop_condition = (df['건축물면적(㎡)'] == 0) & \
                     (df['높이(m)'] == 0) & \
                     (df['데이터생성변경일자'].isna())
    df = df[~drop_condition]
    
    # ---------------------------------------------------------
    # 2. 높이(m)가 0인데 지상층_수가 0이 아닌 행: 높이 = 3 * 지상층_수
    # ---------------------------------------------------------
    mask_h_from_s = (df['높이(m)'] == 0) & (df['지상층_수'] != 0)
    df.loc[mask_h_from_s, '높이(m)'] = 3 * df.loc[mask_h_from_s, '지상층_수']
    
    # ---------------------------------------------------------
    # 3. 건축물면적이 0이고 연면적과 지상층_수가 0이 아닌 경우: 건축물면적 = 연면적 / 지상층_수
    # ---------------------------------------------------------
    mask_a_from_gs = (df['건축물면적(㎡)'] == 0) & \
                     (df['연면적'] != 0) & \
                     (df['지상층_수'] != 0)
    df.loc[mask_a_from_gs, '건축물면적(㎡)'] = df.loc[mask_a_from_gs, '연면적'] / df.loc[mask_a_from_gs, '지상층_수']

    # ---------------------------------------------------------
    # 4. 높이와 지상층_수가 0이고 연면적, 건축물면적이 0이 아닌 경우:
    #    지상층_수 = 연면적 / 건축물면적, 높이 = 지상층_수 * 3
    # ---------------------------------------------------------
    mask_s_h_calc = (df['높이(m)'] == 0) & \
                    (df['지상층_수'] == 0) & \
                    (df['연면적'] != 0) & \
                    (df['건축물면적(㎡)'] != 0)
    
    df.loc[mask_s_h_calc, '지상층_수'] = df.loc[mask_s_h_calc, '연면적'] / df.loc[mask_s_h_calc, '건축물면적(㎡)']
    df.loc[mask_s_h_calc, '높이(m)'] = df.loc[mask_s_h_calc, '지상층_수'] * 3
    
    # ---------------------------------------------------------
    # 5. 높이(m)가 0이고 지하층_수가 0인 경우
    # ---------------------------------------------------------
    mask_logic5 = (df['높이(m)'] == 0) & (df['지하층_수'] == 0)

    df = df[~mask_logic5]
    
    return df
```

```python
cleaned_gdf = clean_building_data(gdf)
```

```python
import pandas as pd
import geopandas as gpd

# 1. 데이터 불러오기
intersected_gdf = gpd.read_file('/Users/judetak/Desktop/산공카르텔경진대회/교차영역_건물.gpkg')
original_grid_gdf = gpd.read_file('/Users/judetak/Desktop/산공카르텔경진대회/격자_고도 포함.gpkg')

# ----------------- [가장 중요한 핵심 해결책] -----------------
# 면적 계산(m²)과 공간 데이터 일치화를 위해 두 레이어 모두 미터 단위인 EPSG:5186으로 변환합니다.
intersected_gdf = intersected_gdf.to_crs(epsg=5186)
original_grid_gdf = original_grid_gdf.to_crs(epsg=5186)
# -----------------------------------------------------------

# 안전장치 1: 병합 키(CELL_ID) 문자열 통일 및 공백 제거
intersected_gdf['CELL_ID'] = intersected_gdf['CELL_ID'].astype(str).str.strip()
original_grid_gdf['CELL_ID'] = original_grid_gdf['CELL_ID'].astype(str).str.strip()

# 안전장치 2: 높이를 강제로 숫자로 변환 (결측치는 0으로)
intersected_gdf['높이(m)'] = pd.to_numeric(intersected_gdf['높이(m)'], errors='coerce').fillna(0)

# 2. 잘려진 건물 조각들의 실제 면적 새로 계산 (이제 미터 단위이므로 정상 계산됨!)
intersected_gdf['piece_area'] = intersected_gdf.geometry.area

print("면적 계산 샘플 (0이 아닌 정상 수치인지 확인):")
print(intersected_gdf[['CELL_ID', 'piece_area', '높이(m)']].head())

# 3. CELL_ID를 기준으로 그룹화하여 합계 계산
grouped = intersected_gdf.groupby('CELL_ID').agg({
    'piece_area': 'sum', 
    '높이(m)': 'sum'      
}).reset_index()

grouped = grouped.rename(columns={'piece_area': 'sum_area', '높이(m)': 'sum_height'})

# 4. 원본 격자에 병합 (Left Join)
final_grid_gdf = pd.merge(original_grid_gdf, grouped, on='CELL_ID', how='left')

# 5. 빈 격자 0으로 채우기
final_grid_gdf['sum_area'] = final_grid_gdf['sum_area'].fillna(0)
final_grid_gdf['sum_height'] = final_grid_gdf['sum_height'].fillna(0)

# 6. 최종 확인
print("\n최종 데이터 샘플:")
print(final_grid_gdf[['CELL_ID', 'sum_area', 'sum_height']].head(10))

# 7. 내보내기
final_grid_gdf.to_file('/Users/judetak/Desktop/산공카르텔경진대회/grid_with_building_stats.gpkg', driver='GPKG', encoding='utf-8')
```

```python
# [수정된 6번 확인 코드] 건물이 존재하는 격자만 필터링해서 보기
valid_grids = final_grid_gdf[final_grid_gdf['sum_area'] > 0]

print("\n--- 🏢 건물이 존재하는 격자 샘플 ---")
print(valid_grids[['CELL_ID', 'sum_area', 'sum_height']].head(10))

print(f"\n총 격자 수: {len(final_grid_gdf)}개")
print(f"건물이 있는 격자 수: {len(valid_grids)}개")
```
