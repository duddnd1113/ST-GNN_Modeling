```python
# 1. 각 잘려진 폴리곤(조각)의 면적 계산
# 주의: 면적을 정확히 계산하려면 gdf의 좌표계가 미터(m) 단위인 투영좌표계(예: EPSG:5179)여야 합니다.
gdf['area'] = gdf.geometry.area

# 2. 피벗 테이블을 사용하여 CELL_ID별, reclass별 면적 합계 구하기
pivot_df = gdf.pivot_table(
    index='CELL_ID', 
    columns='reclass', 
    values='area', 
    aggfunc='sum', 
    fill_value=0  # 해당 피복이 없는 경우 0으로 채움
)

# 3. 각 그리드의 전체 면적 계산 (조각난 면적들의 총합)
pivot_df['total_area'] = pivot_df.sum(axis=1)

# 4. 재분류된 4가지 항목을 퍼센트(%)로 변환 (0 ~ 100%)
target_classes = ['buildings', 'greenspace', 'road struc', 'river zone']

for col in target_classes:
    if col in pivot_df.columns:
        # 해당 피복 면적 / 전체 면적 * 100
        pivot_df[col] = (pivot_df[col] / pivot_df['total_area']) * 100
    else:
        # 데이터 전체에 해당 피복이 아예 없을 경우 0 칼럼 생성
        pivot_df[col] = 0.0

# 퍼센트 계산 후 불필요한 전체 면적 칼럼은 제거
pivot_df = pivot_df.drop(columns=['total_area']).reset_index()

# 5. 원래 그리드가 가지고 있던 고유 정보들(CELL_X, CELL_Y, GID, lon, lat) 추출
# 같은 CELL_ID를 가진 행들 중 첫 번째 행의 정보만 가져와서 중복 제거
grid_info = gdf[['CELL_ID', 'CELL_X', 'CELL_Y', 'GID', 'lon', 'lat']].drop_duplicates(subset=['CELL_ID'])

# 6. 조각난 Geometry(도형)를 다시 하나의 그리드 사각형으로 합치기 (Dissolve)
# 잘려진 조각들을 CELL_ID 기준으로 다시 병합하여 원래의 네모난 그리드 형상으로 복구합니다.
grid_geom = gdf[['CELL_ID', 'geometry']].dissolve(by='CELL_ID').reset_index()

# 7. 속성 정보, 면적 비율(%), 도형(Geometry)을 모두 하나로 병합(Merge)
# grid_info + 퍼센트 데이터
final_df = pd.merge(grid_info, pivot_df, on='CELL_ID', how='left')
# + 도형 데이터
final_df = pd.merge(final_df, grid_geom, on='CELL_ID', how='left')

# 8. 최종 결과를 GeoDataFrame으로 변환 및 원하는 칼럼 순서로 정리
final_gdf = gpd.GeoDataFrame(final_df, geometry='geometry', crs=gdf.crs)

# 요구하신 칼럼 구성(순서)대로 맞추기
columns_order = [
    'CELL_ID', 'buildings', 'greenspace', 'road struc', 'river zone', 
    'CELL_X', 'CELL_Y', 'GID', 'lon', 'lat', 'geometry'
]
final_gdf = final_gdf[columns_order]

# 결과 확인
print(final_gdf.head())
```

```python
import math

# 1. 관측소 데이터 불러오기 및 GeoDataFrame 변환
stations_df = pd.read_csv('/Users/judetak/Desktop/산공카르텔경진대회/seoul_stations.csv')

# 관측소를 구분할 고유 ID 칼럼 설정 (이름이 '측정소명' 등이라면 해당 칼럼으로 변경하세요)
# 만약 고유 ID가 없다면 인덱스를 ID로 만듭니다.
station_id_col = 'station_name' 
if station_id_col not in stations_df.columns:
    stations_df['station_id'] = stations_df.index
    station_id_col = 'station_id'

# dmX(경도), dmY(위도)를 이용해 WGS84(EPSG:4326) 포인트 지오메트리 생성
stations_gdf = gpd.GeoDataFrame(
    stations_df,
    geometry=gpd.points_from_xy(stations_df['dmY'], stations_df['dmX']), # 순서 교체! (dmY=경도, dmX=위도)
    crs="EPSG:4326"
)
# 2. 미터 단위 좌표계(KOREA 2000 / EPSG:5179)로 변환 (버퍼를 m 단위로 치기 위함)
stations_gdf = stations_gdf.to_crs(epsg=5179)

# ⚠️ 주의: 기존 토지피복 데이터도 교차 연산을 위해 동일한 좌표계로 맞춰져 있어야 합니다.
landcover_gdf = gdf.to_crs(epsg=5179)

# 3. 버퍼 거리 및 타겟 변수 설정
buffer_distances = [100, 500, 1000, 2000]
target_classes = ['greenspace', 'river zone', 'road struc', 'buildings']

# 각 버퍼 거리별 결과를 담을 리스트
buffer_results = []

for dist in buffer_distances:
    print(f"{dist}m 버퍼 연산 중...")
    
    # 3-1. 관측소 기준 원형 버퍼 생성
    buffered_stations = stations_gdf.copy()
    buffered_stations.geometry = buffered_stations.buffer(dist)
    
    # 원형 버퍼의 이론적 전체 면적 (원주율 * r^2)
    buffered_stations['buffer_area'] = math.pi * (dist ** 2)
    
    # 3-2. 버퍼 영역과 토지피복도 교차 연산 (Intersect)
    # 버퍼 안으로 들어오는 토지피복 폴리곤만 쿠키틀처럼 잘라냅니다.
    intersected = gpd.overlay(buffered_stations, landcover_gdf, how='intersection')
    
    # 3-3. 잘려진 토지피복의 실제 면적 계산
    intersected['intersect_area'] = intersected.geometry.area
    
    # 3-4. 관측소별, 피복(reclass)별 면적 합계 구하기 (Pivot)
    pivot_df = intersected.pivot_table(
        index=station_id_col,
        columns='reclass',
        values='intersect_area',
        aggfunc='sum',
        fill_value=0
    )
    
    # 3-5. 전체 버퍼 면적 대비 비율(%) 계산
    for col in target_classes:
        if col in pivot_df.columns:
            # 관측소별 전체 버퍼 면적으로 나누기
            buffer_areas = buffered_stations.set_index(station_id_col)['buffer_area']
            pivot_df[col] = (pivot_df[col] / buffer_areas) * 100
        else:
            pivot_df[col] = 0.0  # 해당 반경 내에 그 피복이 아예 없으면 0
            
    # 칼럼 이름에 버퍼 거리 명시 (예: 100m_greenspace)
    pivot_df = pivot_df[target_classes] # 지정한 4개 칼럼만 남기고 순서 정렬
    pivot_df.columns = [f"{dist}m_{col}" for col in pivot_df.columns]
    
    buffer_results.append(pivot_df)

# 4. 모든 버퍼 반경(100, 500, 1000, 2000)의 결과를 하나로 가로 병합
all_buffers_df = pd.concat(buffer_results, axis=1).reset_index()

# 5. 원본 관측소 데이터(위경도 등 포함)와 최종 병합
final_station_df = pd.merge(stations_df, all_buffers_df, on=station_id_col, how='left')

# 버퍼 안에 육지 자체가 없거나 누락된 경우 NaN이 뜰 수 있으므로 0으로 채움
final_station_df = final_station_df.fillna(0)

# 결과 확인
print(final_station_df.head())

```
