### 변수 Feature 뭐뭐 넣을지 시나리오 나누기

```python
from pathlib import Path

# =========================
# 1. 기본 설정
# =========================

id_cols = ['측정소명', '위도', '경도', 'time']

base_transport_cols = [
    '풍향_10m',
    '풍속_10m',
    '동서 방향 풍속',
    '남북 방향 풍속'
]

pm10_mask_cols = [
    'PM10_interp_mask',
    'PM10_spatial_mask'
]

pollutant_cols = [
    'SO2', 'CO', 'O3', 'NO2'
]

pollutant_mask_cols = [
    'SO2_interp_mask', 'CO_interp_mask', 'O3_interp_mask', 'NO2_interp_mask',
    'SO2_spatial_mask', 'CO_spatial_mask', 'O3_spatial_mask', 'NO2_spatial_mask'
]

weather_cols = [
    '기온', '상대습도', '이슬점',
    '현지 기압', '해면 기압'
]

rain_cols = [
    '강수 유무',
    '60분 강수량',
    '일 누적 강수량'
]

# =========================
# 2. mask 요약 변수 생성
# =========================

df_merged['any_interp_mask'] = df_merged[
    ['SO2_interp_mask', 'CO_interp_mask', 'O3_interp_mask', 'NO2_interp_mask', 'PM10_interp_mask']
].max(axis=1)

df_merged['any_spatial_mask'] = df_merged[
    ['SO2_spatial_mask', 'CO_spatial_mask', 'O3_spatial_mask', 'NO2_spatial_mask', 'PM10_spatial_mask']
].max(axis=1)

summary_mask_cols = [
    'any_interp_mask',
    'any_spatial_mask'
]

# =========================
# 3. Feature scenario 정의
# =========================

feature_scenarios = {
    # 기본 transport + PM10
    "S1_transport_pm10": (
        base_transport_cols
        + ['PM10']
    ),

    # PM10 mask만 추가
    "S2_transport_pm10_pm10mask": (
        base_transport_cols
        + ['PM10']
        + pm10_mask_cols
    ),

    # 오염물질 추가, mask 없음
    "S3_transport_pm10_pollutants": (
        base_transport_cols
        + ['PM10']
        + pollutant_cols
    ),

    # 오염물질 추가 + PM10 mask
    "S4_transport_pm10_pollutants_pm10mask": (
        base_transport_cols
        + ['PM10']
        + pollutant_cols
        + pm10_mask_cols
    ),

    # 오염물질 추가 + 전체 오염물질 mask
    "S5_transport_pm10_pollutants_allmask": (
        base_transport_cols
        + ['PM10']
        + pollutant_cols
        + pm10_mask_cols
        + pollutant_mask_cols
    ),

    # 오염물질 추가 + 요약 mask
    "S6_transport_pm10_pollutants_summarymask": (
        base_transport_cols
        + ['PM10']
        + pollutant_cols
        + summary_mask_cols
    ),

    # 기상 변수 추가
    "S7_transport_pm10_weather": (
        base_transport_cols
        + ['PM10']
        + weather_cols
    ),

    # 강수 변수 추가
    "S8_transport_pm10_rain": (
        base_transport_cols
        + ['PM10']
        + rain_cols
    ),

    # 기상 + 강수 추가
    "S9_transport_pm10_weather_rain": (
        base_transport_cols
        + ['PM10']
        + weather_cols
        + rain_cols
    ),

    # 전체 후보 변수 + 요약 mask
    "S10_transport_all_summarymask": (
        base_transport_cols
        + ['PM10']
        + pollutant_cols
        + weather_cols
        + rain_cols
        + ['가시거리']
        + summary_mask_cols
    )
}
```

```python
save_dir = "/home/data/youngwoong/ST-GNN_Dataset/Data_Preprocessed/ST-GNN/feature_scenarios"

import os
os.makedirs(save_dir, exist_ok=True)

for scenario_name, feature_cols in feature_scenarios.items():
    use_cols = id_cols + feature_cols
    use_cols = list(dict.fromkeys(use_cols))

    df_scenario = df_merged[use_cols].copy()
    df_scenario = df_scenario.sort_values(['time', '측정소명']).reset_index(drop=True)

    save_path = f"{save_dir}/{scenario_name}.csv"
    df_scenario.to_csv(save_path, index=False, encoding="utf-8-sig")

    print(f"{scenario_name} 저장 완료")
```

### 일단 먼저 Hold-Out으로 기본 실험만 먼저 해봄.

```python
df_merged = df_merged.sort_values(['time', '측정소명']).reset_index(drop=True)

# 고유 시간 기준 split
times = sorted(df_merged['time'].unique())

n = len(times)
train_end = int(n * 0.7)
valid_end = int(n * 0.85)

train_times = times[:train_end]
valid_times = times[train_end:valid_end]
test_times  = times[valid_end:]

# 실제 데이터 분할
df_train = df_merged[df_merged['time'].isin(train_times)].copy()  # 70퍼
df_valid = df_merged[df_merged['time'].isin(valid_times)].copy()  # 15퍼
df_test  = df_merged[df_merged['time'].isin(test_times)].copy()   # 15퍼

print("전체:", df_merged.shape)
print("Train:", df_train.shape, df_train['time'].min(), "~", df_train['time'].max())
print("Valid:", df_valid.shape, df_valid['time'].min(), "~", df_valid['time'].max())
print("Test :", df_test.shape,  df_test['time'].min(),  "~", df_test['time'].max())
```

```python
from pathlib import Path
import pickle

split_info = {
    "train_times": train_times,
    "valid_times": valid_times,
    "test_times": test_times,
}

save_path = Path("split_info.pkl")

with open(save_path, "wb") as f:
    pickle.dump(split_info, f)

print(f"split 저장 완료: {save_path}")
```
